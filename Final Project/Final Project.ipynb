```python
# Импортируем библиотеки

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns
import json
import pickle
import re
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor

import warnings
warnings.filterwarnings('ignore')
```

Проходим по всем 15 файлам дампа НКО и обьединяем в один дата сет

```python
data ={}
temp = []

# Проходим по всем 15 файлам дампа НКО
for i in range(1,16):
    
    if i < 10:
       path = f'data/openngo_2021_26_08/ngo_dump_0{i}.json'
    else:
        path = f'data/openngo_2021_26_08/ngo_dump_{i}.json' 
        
    # Открываем файл в режиме чтения с UTF-8 кодировкой    
    with open(path, 'r', encoding='utf-8') as f:

            temp = []
            
            for line in f:
                line = line.strip()
                
                if line:
                    obj = json.loads(line)
                    temp.append(obj)
             
            # Сохраняем данные файла в основной словарь с ключом file_X       
            data[f'file_{i}'] = temp

# Создаем пустой DataFrame для объединения всех данных из JSON-файлов
full_df = pd.DataFrame()

for elem in list(data.keys()):
    # Нормализуем JSON-данные из файла в плоский DataFrame
    df = pd.json_normalize(data[elem])
    
    # Конкатенируем текущий DataFrame к общему
    full_df = pd.concat([full_df, df], axis=0)
    full_df = full_df.reset_index(drop=True)
    
full_df.head(3)
```
