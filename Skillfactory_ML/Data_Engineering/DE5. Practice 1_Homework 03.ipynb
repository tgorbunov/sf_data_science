Инструменты: Selenium для автоматизации взаимодействия с сайтом и Beautiful Soup для анализа HTML.

Скринкаст: Практика 1. Парсинг теста с сайта TProger

Импорт библиотек

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

Установка драйверов

```python
# Задаём путь до файла с драйвером для браузера
driver_path = './chromedriver-win64/chromedriver.exe'

# Создаём веб-драйвер для работы

def run(driver_path):
    driver = webdriver.Chrome(service=ChromeService(executable_path=driver_path))
    driver.get("https://tproger.ru/quiz/real-programmer/")
    
    button = driver.find_element_by_class_name("quiz_button")
    button.click()
    
    question = driver.find_element_by_class_name("question").text
    
    answer = driver.find_element_by_xpath('//*[@id="quiz_item_255"]/div[2]/div[1]')
    answer.click()
    
    answers = driver.find_elements_by_class_name("answer ")
    for i in answers:
        if i.text:
            print(i.text)
    
    right_ans=driver.find_element_by_class_name("correct_answer").text
    print('right answer is ', right_ans)
    driver.close()
    driver.quit()
 
run(driver_path)
```
