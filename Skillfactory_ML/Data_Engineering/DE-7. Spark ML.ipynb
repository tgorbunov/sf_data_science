### Чтение данных

PySpark может создавать распределённые наборы данных из любого источника хранилища, поддерживаемого Hadoop, включая вашу локальную файловую систему, HDFS, Cassandra, HBase, Amazon S3 и так далее. Spark поддерживает текстовые файлы, SequenceFiles и любой другой формат ввода Hadoop.

Текстовые файлы RDD можно создавать с помощью метода textFile SparkContext. Этот метод принимает URI для файла (либо локальный путь на машине, либо URI hdfs://, s3a:// и так далее) и считывает его как набор строк. Вот пример вызова:

```python
textRdd =  spark.sparkContext.textFile("README.md")
```

Для чтения данных сразу в dataframe используем другую команду:

```python
df = spark.textFile("/tmp/resources/zipcodes.csv")
```

либо:

```python
df = spark.read.csv("/tmp/resources/zipcodes.csv")
```
