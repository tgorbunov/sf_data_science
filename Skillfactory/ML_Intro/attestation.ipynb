```python
# Расчет RMSE
import numpy as np
from sklearn.metrics import mean_squared_error

# Данные
y_true = [1.23, 2.35, 2.75]
y_pred = [1.01, 12.3, 2.74]

# Способ 1: Вручную
squared_errors = [(true - pred)**2 for true, pred in zip(y_true, y_pred)]
mse = sum(squared_errors) / len(squared_errors)
rmse_manual = np.sqrt(mse)

print("Пошаговый расчет:")
print(f"Квадраты ошибок: {squared_errors}")
print(f"Сумма квадратов: {sum(squared_errors)}")
print(f"MSE: {mse}")
print(f"RMSE (вручную): {rmse_manual}")

# Способ 2: Через sklearn
rmse_sklearn = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"RMSE (sklearn): {rmse_sklearn}")

# Округление до 2 знаков
rmse_rounded = round(rmse_manual, 2)
print(f"\nОтвет: {rmse_rounded}")
```

```python
# Расчет Accuracy для классификатора новостей
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np

print("=== ЗАДАЧА: КЛАССИФИКАЦИЯ НОВОСТЕЙ ===")
print()

# Исходные данные
total_political = 15    # Всего политических новостей
total_economic = 20     # Всего экономических новостей
total_news = total_political + total_economic

print("Исходные данные:")
print(f"Политических новостей: {total_political}")
print(f"Экономических новостей: {total_economic}")
print(f"Всего новостей: {total_news}")
print()

# Ошибки классификатора
political_misclassified = 9  # политические, но помечены как экономические
economic_misclassified = 6   # экономические, но помечены как политические

print("Ошибки классификатора:")
print(f"Политические новости, помеченные как экономические: {political_misclassified}")
print(f"Экономические новости, помеченные как политические: {economic_misclassified}")
print()

# Правильно классифицированные
political_correct = total_political - political_misclassified
economic_correct = total_economic - economic_misclassified
total_correct = political_correct + economic_correct

print("Правильно классифицированные:")
print(f"Политические новости (правильно): {political_correct}")
print(f"Экономические новости (правильно): {economic_correct}")
print(f"Всего правильных: {total_correct}")
print()

# Расчет Accuracy
# Accuracy = (True Positives + True Negatives) / Total
accuracy = total_correct / total_news
accuracy_percent = accuracy * 100

print("=== РАСЧЕТ ACCURACY ===")
print(f"Accuracy = Правильные предсказания / Общее количество")
print(f"Accuracy = {total_correct} / {total_news} = {accuracy:.4f}")
print(f"Accuracy = {accuracy_percent:.2f}%")
print()

# Округление до 2 знаков
accuracy_rounded = round(accuracy, 2)
print(f"ОТВЕТ: {accuracy_rounded}")
print()

# Построение матрицы ошибок для наглядности
print("=== МАТРИЦА ОШИБОК ===")
print("                    Предсказано")
print("                 Политич.  Эконом.")
print(f"Истинно Политич.     {political_correct}        {political_misclassified}")
print(f"        Эконом.      {economic_misclassified}       {economic_correct}")
print()

# Создание данных для проверки через sklearn
# Кодировка: 0 = политические, 1 = экономические
y_true = [0] * total_political + [1] * total_economic
y_pred = ([0] * political_correct + [1] * political_misclassified + 
          [0] * economic_misclassified + [1] * economic_correct)

# Проверка через sklearn
accuracy_sklearn = accuracy_score(y_true, y_pred)
print(f"Проверка через sklearn: {accuracy_sklearn:.4f}")
print(f"Совпадает с ручным расчетом: {abs(accuracy - accuracy_sklearn) < 1e-10}")
```
