```python
fig = plt.figure(figsize=(8, 5))
boxplot = sns.boxplot(
    data=sber_data,
    x='ecology',
    y='price_doc',
    orient='v',
    width=0.5,
    hue='ecology'
)
#boxplot.set_xticklabels(['Лояльные (0)', 'Ушедшие (1)'])
boxplot.set_xlabel('Экология')
boxplot.set_ylabel('Цена квартиры')
boxplot.set_title('Распределение цен квартир по экологии')
boxplot.set(ylim=(0, 20000000))
boxplot.grid()
```

```python
sber_data_price_kremlin = sns.jointplot(
    data=sber_data, 
    x='price_doc', 
    y='kremlin_km',
    hue='ecology',
    xlim=(0, 60000000),
    ylim=(0, 50),
    height=10,
    size = 0.2
)
```
