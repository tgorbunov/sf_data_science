# 2. Линейный оператор

### Линейный оператор

Умножение матрицы на вектор можно представить как его одновременный поворот и растяжение.

А если рассмотреть умножение матрицы на все векторы пространства, то получится преобразование всего этого пространства или так называемый **линейный оператор**.

### Собственные векторы и числа

**Собственный вектор или айгенвектор** — это вектор, который не меняет направление под действием оператора, а только растягивается или сжимается.

Коэффициент растяжения или сжатия λ («лямбда») называется **собственным числом** оператора А и его матрицы.

**Айгенпарой** называется пара, состоящая из собственного числа и соответствующего ему собственного вектора. 

**Спектром матрицы** называется набор её собственных чисел.

# 4. Практика: собственные векторы и числа

В этом юните мы посмотрим, как вычислять собственные векторы и соответствующие им собственные числа с помощью Python.

Для вычисления собственных чисел в библиотеке numpy предусмотрена специальная функция np.linalg.eig(). Эта функция возвращает кортеж, который состоит из собственных чисел и матрицы, составленной из собственных векторов, соответствующей собственным числам (собственные векторы записаны в столбцы матрицы).

Давайте решим с помощью Python примеры, которые мы рассматривали в видео. Но предварительно импортируем библиотеку numpy:

```python
import numpy as np
```

```python
# создаем матрицу А
A = np.array([
    [3, 1],
    [2, 2]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

```python
# создаем матрицу А
A = np.array([
    [1, -5, -6],
    [4, 8, 7],
    [5, 9, 11]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

```python
# создаем матрицу А
A = np.array([
    [1, -4, -5, -6],
    [4, 12, 8, 7],
    [5, 14, 9, 11],    [8, 15, 7, 4]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

```python
# создаем матрицу А
A = np.array([
    [0, 1],
    [-1, 0],
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

Опять магия! Результат есть. Только какой-то странный: что за +1.j и -1.j? Забегая вперёд, скажем, что это **комплексные числа**. Однако оставим эту интригу и разберёмся с ними чуть позже.

```python
from sklearn import datasets 
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
boston_data = pd.read_csv('data/housing.csv', header=None, delimiter=r"\s+", names=column_names)
boston_data.head()
```

Составим матрицу наблюдений, в которую войдут все исходные факторы и вектор целевой переменной: 

```python
A = boston_data.drop('PRICE', axis=1)
y = boston_data[['PRICE']]
```

Затем составим корреляционную матрицу для исходных факторов:

```python
C = A.corr()
fig = plt.figure(figsize=(10, 5))
sns.heatmap(C, annot=True, cmap='coolwarm');
```

А теперь найдём собственные числа полученной корреляционной матрицы:

```python
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(C)
print('Собственные числа: \n', eig_values)
#print('Собственные векторы: \n', eig_vectors)
```

Так как корреляционная матрица симметричная, то её собственные числа всегда действительны и различны между собой. Это мы и наблюдаем. 

Проверим, что все собственные векторы корреляционной матрицы **ортогональны** друг другу. Это свойство понадобится нам в **методе главных компонент**.

```python
# считаем матрицу Грамма L^T*L:
print(np.round(eig_vectors.T @ eig_vectors, 2))
```

⭐️ Та-дам! Мы получили единичную матрицу, а значит, собственные вектора ортогональны друг другу. Мы доказали ещё одно свойство собственных чисел для симметричных матриц.

А теперь перейдём к практическим заданиям.

```python
# создаем матрицу А
A = np.array([
    [1, 4, 13],
    [3, -4, 7],
    [5, 9, 12]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', np.round(eig_values, 0))
```

```python
# создаем матрицу А
A = np.array([
    [1, 9, 4],
    [9, 4, 7],
    [4, 7, 12]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A.T@A)
print('Собственные числа: \n', np.round(eig_values, 0))
```
